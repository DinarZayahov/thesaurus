```python
from thesaurus import Thesaurus
```

```python
obj = Thesaurus(lang='eng')
```

```python
background1 = obj.read_pickle('../data/2010')
background2 = obj.read_pickle('../data/2017')
background3 = obj.read_pickle('../data/2020')
text1 = obj.read_txt('../data/shakespeare.txt')
```

```python
# texts is the dictionary with all source texts that will be preprocessed
# texts = {'foreground name1': [source text1, source text2, ...], ...}
texts = dict()

# list of foreground names
# foreground_names = ['foreground name1', ...]
foreground_names = []

foreground_name1 = 'Lover\'s Complaint by William Shakespeare'
texts[foreground_name1] = [text1]
foreground_names.append(foreground_name1)
```

```python
MODEL = 'en_core_web_md-3.0.0/en_core_web_md/en_core_web_md-3.0.0'
obj.set_spacy_model(MODEL)
```

```python
# dictionary with the embeddings and tokens of each foreground unit
# processed_foregrounds = {'foreground_name1': {'embeds': embeddings, 'words': tokens}, ...}
processed_foregrounds = obj.process_foreground(foreground_names, texts)
```

```python
background_embeds, background_words = obj.process_texts(obj.custom_preprocessing_of_data(background1) + obj.custom_preprocessing_of_data(background2) + obj.custom_preprocessing_of_data(background3))
```

```python
obj.set_som('train', background_embeds)
```

```python
print("Go to https://nbviewer.org/github/DinarZayahov/thesaurus/blob/master/ThesaurusVisualization.ipynb to see the outputs of the notebook as the GitHub doesn't render dynamic output")
obj.show_map(background_embeds, background_words, foreground_names, processed_foregrounds, model='new')
```
